Practical 7.	Measurement of seasonal indices Link Relative method




```python
import numpy as np
import pandas as pd

# Step 1: Input the quarterly sales data
data = {
    'Year': [2022, 2023, 2024],
    'Q1': [200, 210, 220],
    'Q2': [250, 260, 270],
    'Q3': [180, 190, 200],
    'Q4': [220, 230, 240]
}
df = pd.DataFrame(data)

# Step 2: Compute Link Relatives
link_relatives = []
for i in range(len(df)):
    row = []
    if i > 0:
        row.append((df.loc[i, 'Q1'] / df.loc[i - 1, 'Q4']) * 100)  # Q1 vs Previous Q4
    for j in range(1, 4):
        row.append((df.iloc[i, j] / df.iloc[i, j - 1]) * 100)  # Other quarters
    link_relatives.append(row)

# Convert to DataFrame
link_rel_df = pd.DataFrame(link_relatives, columns=['Q1', 'Q2', 'Q3', 'Q4'])

# Step 3: Calculate Average Link Relatives
avg_link_relatives = link_rel_df.mean()

# Step 4: Adjust Seasonal Indices
sum_avg = avg_link_relatives.sum()
adjustment_factor = 400 / sum_avg
seasonal_indices = avg_link_relatives * adjustment_factor

# Display Results
print("Average Link Relatives:")
print(avg_link_relatives)
print("\nAdjusted Seasonal Indices:")
print(seasonal_indices)
```
