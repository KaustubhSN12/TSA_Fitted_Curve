# 3.	Fitting and plotting of logistic curve

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Generate synthetic data
np.random.seed(42)
x_data = np.linspace(0, 10, 100)
y_data = logistic(x_data, L=10, k=1, x0=5) + np.random.normal(0, 0.5, size=len(x_data))

# Fit logistic curve to data
popt, _ = curve_fit(logistic, x_data, y_data, p0=[10, 1, 5])
L_fit, k_fit, x0_fit = popt

# Generate fitted curve
x_fit = np.linspace(0, 10, 100)
y_fit = logistic(x_fit, L_fit, k_fit, x0_fit)

# Plot results
plt.scatter(x_data, y_data, label='Data', alpha=0.6)
plt.plot(x_fit, y_fit, color='red', label=f'Fitted Logistic Curve\nL={L_fit:.2f}, k={k_fit:.2f}, x0={x0_fit:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Logistic Curve Fitting')
plt.show()
```
