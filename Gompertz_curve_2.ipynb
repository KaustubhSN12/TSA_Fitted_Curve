###Practical 2: Fitting and Plotting of Gompertz Curve

The Gompertz curve is used for modeling growth processes like population growth, tumor growth, and business forecasting.

Mathematical Form
The general form of the Gompertz function is:

$$y(t) = a \cdot \exp(-b \cdot \exp(-c \cdot t))$$


Parameters:

$a$: The upper asymptote (maximum value).

$b$: The displacement parameter.

$c$: The growth rate.

$t$: The independent variable (e.g., time).

# Steps to Fit and Plot
Generate or load data.

Define the Gompertz function.

Fit the function to data using scipy.optimize.curve_fit.

Plot the fitted curve along with the data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the Gompertz function
def gompertz(x, a, b, c):
    return a * np.exp(-b * np.exp(-c * x))

# Generate synthetic data (for demonstration)
np.random.seed(42)
x_data = np.linspace(0, 10, 100)  # Time points
a_true, b_true, c_true = 100, 2, 0.5  # True parameters

y_data = gompertz(x_data, a_true, b_true, c_true) + np.random.normal(0, 2, size=len(x_data)) # Add noise

# Fit the Gompertz function to the data
popt, pcov = curve_fit(gompertz, x_data, y_data, p0=[100, 2, 0.5])  # Initial guess

# Extract fitted parameters
a_fit, b_fit, c_fit = popt
print(f"Fitted parameters: a={a_fit:.2f}, b={b_fit:.2f}, c={c_fit:.2f}")

# Plot the data and fitted curve
plt.scatter(x_data, y_data, label="Data", color='blue', alpha=0.6)
plt.plot(x_data, gompertz(x_data, *popt), label="Fitted Gompertz Curve", color='red')
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()
plt.title("Gompertz Curve Fitting")
plt.show()
```
