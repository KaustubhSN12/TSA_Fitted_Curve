#P4 :  Fitting if trend by Moving Average method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example data : monthly sales over 24 months
data =  [120 ,150 , 170, 200,250, 270,300, 320, 350, 400, 450, 470,
         130, 160 ,180, 210, 260, 280, 310, 330, 360, 410, 460, 480]

months = pd.date_range(start="2022-01", periods= 24, freq="M")

#create Dataframe
df = pd.DataFrame({'Date': months, 'Sales': data})
df.set_index('Date', inplace=True)

#moving Average with a 12-month window
df['12-Month MA'] = df ['Sales'].rolling(window=12).mean()

```

```python
#Plot original  Data and Trend
plt.figure(figsize=(12,6))
plt.plot(df['Sales'],label='Original Sales', marker='o')
plt.plot(df['12-Month MA'], label='12-Month Moving Average (Trend)',color ='red')
plt.title('fitting Trend using Moving Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```
