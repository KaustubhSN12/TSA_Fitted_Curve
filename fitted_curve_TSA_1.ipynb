###Practical 1 Fitting and plotting of modified exponential curve

Decreaseing Curve

popt (Optimal Parameters) What it is:An array holding the specific numbers (coefficients) for your chosen function (e.g., a, b, c in $ax^{2}+bx+c$) that best fit the provided data points.

**Purpose**: To define the equation of the fitted curve.

**pcov** (Parameter Covariance) What it is: A square matrix showing how much the fitted parameters change together.

**Diagonal Elements**: The variance (squared standard error) of each parameter.

Off-Diagonal Elements **bold text**: The covariance between pairs of parameters, indicating their interdependency (e.g., if the intercept goes up, the slope must go down).


How to use it **bold text**: You take the square root of the diagonal elements to get the standard error (uncertainty) for each parameter (e.g., np.sqrt(np.diag(pcov))).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the modified exponential model: y(x) = a * exp(-b * x) + c
def model(x, a, b, c):
    return a * np.exp(-b * x) + c

# Example: Generate synthetic data with noise
x_data = np.linspace(0, 10, 50) # Independent variable (e.g., time)
# True parameters: a=5, b=0.8, c=2
y_data = 5 * np.exp(-0.8 * x_data) + 2 + np.random.normal(0, 0.5, len(x_data)) # adding noise

# Fit the model to the data
popt, pcov = curve_fit(model, x_data, y_data, p0=[1, 1, 1])
# Initial guess for parameters

# Extract fitted parameters
a_fit, b_fit, c_fit = popt

# Print the fitted parameters
print(f"Fitted parameters: a = {a_fit:.3f}, b = {b_fit:.3f}, c = {c_fit:.3f}")

# Generate fitted curve data
y_fit = model(x_data, *popt)

# Plot the data and fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, color='blue', label='Data', alpha=0.6)
plt.plot(x_data, y_fit, color='red', label=f'Fitted Curve: a={a_fit:.2f}, b={b_fit:.2f}, c={c_fit:.2f}')
plt.title('Fitting of Modified Exponential Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
