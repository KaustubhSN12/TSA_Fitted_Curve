**Practical 8. Calculation of variance of random component by variate difference method**

```python
import numpy as np

def variate_difference_variance(data):
    """
    Calculate the variance of a random component using the variate difference method.

    Parameters:
        data (list or numpy array): A sequence of observations.

    Returns:
        float: Estimated variance.
    """
    n = len(data)
    if n < 2:
        raise ValueError("Data must contain at least two observations.")

    diff_sq = [(data[i+1] - data[i])**2 for i in range(n-1)]
    variance = (1 / (2 * (n - 1))) * sum(diff_sq)

    return variance

# Example usage
data = [10, 12, 15, 14, 18, 20, 19]  # Example dataset
variance = variate_difference_variance(data)
print(f"Estimated variance: {variance:.4f}")
```
