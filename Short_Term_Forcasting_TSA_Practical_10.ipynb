**Practical 10.Forecasting by short term forecasting methods**

**Short-term forecasting methods in Python typically include techniques like:**

Moving Average (MA) – Simple and weighted moving averages.


Exponential Smoothing (SES, DES, TES) – Simple, Double, and Triple Exponential Smoothing.


ARIMA (AutoRegressive Integrated Moving Average) – A statistical model for time series forecasting.


SARIMA (Seasonal ARIMA) – ARIMA with seasonal components.


Prophet – Developed by Facebook for time series forecasting.


LSTM (Long Short-Term Memory) – A deep learning approach for sequential data.

```python
#1. Moving Average (MA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
data = np.cumsum(np.random.randn(100))  # Random walk time series

# Moving average
window = 5
moving_avg = pd.Series(data).rolling(window=window).mean()

# Plot
plt.plot(data, label="Actual")
plt.plot(moving_avg, label=f"Moving Average (Window={window})", linestyle="dashed")
plt.legend()
plt.show()
```

```python
#2. Exponential Smoothing

from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Fit model
ses_model = SimpleExpSmoothing(data).fit(smoothing_level=0.2, optimized=False)
forecast = ses_model.forecast(10)

# Plot
plt.plot(data, label="Actual")
plt.plot(ses_model.fittedvalues, label="SES Fitted", linestyle="dashed")
plt.plot(range(len(data), len(data) + 10), forecast, label="Forecast", linestyle="dotted")
plt.legend()
plt.show()
```

```python
#3. ARIMA
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(data, order=(2,1,2))
arima_fit = model.fit()

# Forecast
forecast = arima_fit.forecast(steps=10)

# Plot
plt.plot(data, label="Actual")
plt.plot(range(len(data), len(data) + 10), forecast, label="Forecast", linestyle="dotted")
plt.legend()
plt.show()
```
